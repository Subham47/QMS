```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('tilt.csv')
df = pd.DataFrame(df, columns=["Days", "Monthly","Seasonal","Semi_yearly","Yearly"])
```

```python
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
df1 = scaler.fit_transform(np.array(df1).reshape(-1,1))
```

```python
df1
```

```python
training_size = int(len(df1)*0.70)
test_size = len(df1) - training_size
train_data, test_data = df1[0:training_size, :],df1[training_size:len(df1), :1]
```

```python
training_size,test_size
```

```python
import numpy 
def create_dataset(dataset, time_step= 1):
  dataX, dataY = [], []
  for i in range(len(dataset)-time_step-1):
    a= dataset[i:(i+time_step), 0]
    dataX.append(a)
    dataY.append(dataset[i + time_step, 0])
  return numpy.array(dataX), numpy.array(dataY)
```

```python
time_step= 10
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

```python
print(X_train.shape),print(y_train.shape), print(X_test.shape), print(y_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
print(X_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True,input_shape=(7,1)))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss= 'mean_squared_error' ,optimizer= 'adam' )
```

```python
from timeit import default_timer as timer
epochs = 100

start = timer()  
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=8,verbose=1)
print("with CPU time taken in seconds:", timer()-start) 
```

```python
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
math.sqrt(mean_squared_error(y_test,test_predict))
```

```python
look_back=10
trainPredictPlot = numpy.empty_like(df1)
trainPredictPlot[:,:] =np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict

testPredictPlot = numpy.empty_like(df1)
testPredictPlot[:, :] = numpy.nan
random.uniform(0, 1)
testPredictPlot[len(train_predict)+(look_back*2)+1:len(df1)-1, :] = test_predict

plt.figure(figsize=(10,5))
plt.plot(scaler.inverse_transform(df1))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.xlabel("data number")
plt.ylabel("efficiency")
plt.title("training and testing plot")
plt.show()
```
